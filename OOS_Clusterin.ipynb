```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import operator
import json

embedder = SentenceTransformer('all-MiniLM-L6-v2') #- pre-trained models 
#embedder=SentenceTransformer('multi-qa-distilbert-cos-v1')


import pandas as pd
#file intake
url = '/content/sample_Test.csv'
df = pd.read_csv(url)
corpus=df['utterances']


corpus_embeddings = embedder.encode(corpus)

# Normalize the embeddings to unit length
corpus_embeddings = corpus_embeddings /  np.linalg.norm(corpus_embeddings, axis=1, keepdims=True)

# Perform  clustering
 #, affinity='cosine', linkage='average', distance_threshold=0.4)
clustering_model = AgglomerativeClustering(n_clusters=None, distance_threshold=1.0, linkage="ward")
clustering_model.fit(corpus_embeddings)
cluster_assignment = clustering_model.labels_

print("Start clustering")

clustered_sentences = {}
for sentence_id, cluster_id in enumerate(cluster_assignment):
    if cluster_id not in clustered_sentences:
        clustered_sentences[cluster_id] = []

    clustered_sentences[cluster_id].append(corpus[sentence_id])


sorted_group = sorted(clustered_sentences.items(), key=operator.itemgetter(0))
 

data_OOS = pd.DataFrame(sorted_group)
data_OOS.columns=['Cluster', 'Utterances']
data_OOS.head()
data_OOS.to_json('cluster_oos.json', orient='records')
```
